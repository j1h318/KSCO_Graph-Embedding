## skill node 추가

```python
path = '/content/drive/MyDrive/balanceup/content/skills.csv'
df = pd.read_csv(path)
jobs = df.values.tolist()
result = list()
for job in jobs:
    occu3 = str(job[0])
    jobsCd = str(job[2])
    if len(occu3) == 5:
        occu3 = "0" + occu3
    if len(jobsCd) == 5:
        jobsCd = "0" + jobsCd

    skill = job[3].replace("_", ",")
    skill = skill.replace('학력',"")
    skill = skill.replace('경력',"")
    skill = skill.replace('응시자격',"")
    skill = skill.replace('지역거주자',"")
    skill = skill.replace('사설학원',"")
    skills = skill.replace('독학',"")
    skills = skills.lower()
    skills = skills.split(',')
    ln = len(skills)
    for i in range(ln-1,-1,-1):
        if skills[i] == " ":
            skills.pop(i)
    skills = [tok.strip() for tok in skills]
    for sk in skills:
      if not sk in result:
        result.append(sk)
result
```

```python
for sk in result:
  doc = "CREATE (skill : SKILL { skill : '" + sk +"'})"
  session.run(doc)
  print(doc)
```

## occupation, jaccard relation add

```python
path = '/content/drive/MyDrive/balanceup/content/jobs_jaccard.csv'
df = pd.read_csv(path)
row = df.shape[0]
jaccard_rel = dict()
for i_idx in range(row):
  st_occu = str(df.loc[i_idx]['기준 직업 코드'])
  com_occu = str(df.loc[i_idx]['비교 직업 코드'])
  jaccard = df.loc[i_idx]['jaccard']
  if len(st_occu) == 5:
    st_occu = "0" + st_occu
  if len(com_occu) == 5:
    com_occu = "0" + com_occu
  if jaccard_rel.get(st_occu+com_occu) is None:
    jaccard_rel[st_occu+com_occu] = [st_occu, com_occu, jaccard]
print(len(jaccard_rel))

```

```python
keys = list(jaccard_rel.keys())
for key in keys:
  query = "MATCH (o1:OCCUPATION{ occupationCd : '"+ jaccard_rel[key][0]  + "'}), (o2:OCCUPATION{ occupationCd :'" + jaccard_rel[key][1] +"'})" + " CREATE (o1)-[j:JACCARD]->(o2)" + "SET j.distance = ' " + str(jaccard_rel[key][2]) + "'"
  print(query)
  session.run(query)
```
